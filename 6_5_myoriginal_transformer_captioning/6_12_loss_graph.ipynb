```python
import matplotlib.pyplot as plt

f1 = open( '../model/6-12_train_loss_20240224_142016.csv')
f2 = open( '../model/6-12_val_loss_20240224_142016.csv')

line1 = f1.readline()
line1 = line1.replace( "\n", "")

line2 = f2.readline()
line2 = line2.replace( "\n", "")


i = 0
x = []
loss = []
v_i = 0
v_x = []
v_loss = []

while line1:
    #print( line1 )
    line1_split = line1.split(",")
    #print( line1_split[0])
    i += 1
    x.append( i / 580 + 1 )
    loss.append( float(line1_split[1].split(' ')[1]) )
    line1 = f1.readline()
    line1 = line1.replace( "\n", "")
    
while line2:
    #print( line1 )
    line2_split = line2.split(",")
    #print( line1_split[0])
    v_i += 1
    v_x.append( v_i / 249 + 1 )
    v_loss.append( float(line2_split[1].split(' ')[1]) )
    line2 = f2.readline()
    line2 = line2.replace( "\n", "")
   
    
plt.plot( x, loss)
plt.title( "Train Loss")
plt.xlabel( 'Epochs')
plt.ylabel( 'Loss')

plt.show()

plt.plot( v_x, v_loss)
plt.title( "Validation Loss")
plt.xlabel( 'Epochs')
plt.ylabel( 'Loss')

plt.show()
```
